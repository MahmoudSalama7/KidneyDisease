```python
import random
from PIL import Image

def display_sample_images(paths, categories):
    plt.figure(figsize=(12, 8))
    
    for i, (path, category) in enumerate(zip(paths, categories)):
        img_name = random.choice(os.listdir(path))
        img_path = os.path.join(path, img_name)
        
        img = Image.open(img_path)
        
        plt.subplot(2, 2, i+1)
        plt.imshow(img)
        plt.title(f'{category} Sample Image')
        plt.axis('off')
    
    plt.tight_layout()
    plt.show()

# Define the paths for each category
paths = [path_cyst, path_tumor, path_stone, path_normal]
display_sample_images(paths, categories)
```

```python
def get_image_sizes(paths):
    image_sizes = {}
    
    for path, category in zip(paths, categories):
        img_name = random.choice(os.listdir(path))
        img_path = os.path.join(path, img_name)
        img = Image.open(img_path)
        image_sizes[category] = img.size
    
    return image_sizes

image_sizes = get_image_sizes(paths)
image_sizes
```

```python
def check_corrupted_images(paths):
    corrupted_files = []
    
    for path in paths:
        for img_name in os.listdir(path):
            img_path = os.path.join(path, img_name)
            try:
                img = Image.open(img_path)
                img.verify()  # Verify that the file is an image
            except (IOError, SyntaxError) as e:
                corrupted_files.append(img_path)
    
    return corrupted_files

corrupted_images = check_corrupted_images(paths)
print(f"Number of corrupted images: {len(corrupted_images)}")
```

```python
def collect_image_dimensions(paths):
    width_list, height_list = [], []
    
    for path in paths:
        for img_name in os.listdir(path):
            img_path = os.path.join(path, img_name)
            with Image.open(img_path) as img:
                width, height = img.size
                width_list.append(width)
                height_list.append(height)
    
    return width_list, height_list

widths, heights = collect_image_dimensions(paths)

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
sns.histplot(widths, kde=True)
plt.title('Image Width Distribution')

plt.subplot(1, 2, 2)
sns.histplot(heights, kde=True)
plt.title('Image Height Distribution')

plt.tight_layout()
plt.show()
```

The Image Width Distribution reveals that the majority of the images in the dataset have a width around 400 pixels, with a significant spike at this value. While there are some images with larger widths, ranging up to 1200 pixels, they are much fewer in number, creating a right-skewed distribution. This means most images are relatively small, with only a minority having larger widths.

Similarly, the Image Height Distribution shows a dominant concentration of images between 500-700 pixels in height. Like the width distribution, there are fewer images with larger heights, with some reaching up to 1100 pixels. The overall trend is again right-skewed, indicating that the majority of images are clustered at smaller heights, with fewer larger ones.

These patterns suggest that the images in the dataset are mostly of moderate size, and resizing them to a consistent dimension (like 200x200 pixels, used in your model) would help standardize input for the neural network. The presence of some larger images also implies that proper resizing, cropping, or normalization will be essential to maintain a uniform input structure.

```python
train_datagen = ImageDataGenerator(rescale=1/255)
valid_datagen = ImageDataGenerator(rescale=1/255)
test_datagen = ImageDataGenerator(rescale=1/255)
```

```python
train_dataset = train_datagen.flow_from_directory('/kaggle/input/kidney-diseases-recognition/dataset',
                                                  target_size=(200, 200),
                                                  color_mode='grayscale', 
                                                  class_mode='categorical', 
                                                  batch_size=100,
                                                  )

```

```python
model = tf.keras.models.Sequential([
  
    tf.keras.layers.Conv2D(32,(3,3),strides=(1, 1),activation='relu',padding='same', input_shape=train_dataset.image_shape), 
    tf.keras.layers.MaxPooling2D(pool_size=(2,2)),
    
    tf.keras.layers.Conv2D(64,(3,3),strides=(1, 1) ,padding='same',activation='relu'),
    tf.keras.layers.MaxPooling2D(pool_size=(2,2)),

    tf.keras.layers.Conv2D(128,(3,3),strides=(1, 1),padding='same', activation='relu'),
    tf.keras.layers.MaxPooling2D(pool_size=(2,2)),

    tf.keras.layers.Conv2D(256,(3,3),strides=(1, 1),padding='same', activation='relu'),
    tf.keras.layers.MaxPooling2D(pool_size=(2,2)),


    tf.keras.layers.Flatten(),

    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(64, activation='relu'),  # hidden layers 
    
    
    tf.keras.layers.Dense(4, activation='softmax'),
 
])
```

```python
model.summary()
```

```python
import keras
METRICS = [
        'accuracy',
        keras.metrics.Precision(name='precision'),
        keras.metrics.Recall(name='recall')
    ]
    
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=METRICS)
```

```python
model_fit = model.fit(train_dataset,
                      validation_data=valid_dataset,
                      steps_per_epoch = 3,
                      epochs = 30)
```

```python
fig, ax = plt.subplots(1, 4, figsize=(20, 5))
ax = ax.ravel()

for i, met in enumerate(['precision', 'recall', 'accuracy', 'loss']):
    ax[i].plot(model_fit.history[met])
    ax[i].plot(model_fit.history['val_' + met])
    ax[i].set_title('Model {}'.format(met))
    ax[i].set_xlabel('epochs')
    ax[i].set_ylabel(met)
    ax[i].legend(['train', 'val'])
```

## Version 1: Initial Model (16 Filter with 3 Layers)
I started by building a very basic model to quickly evaluate the dataset and establish a baseline. At this stage, I used 3 dense layers and no convolutional layers, which was overly simplistic but good for an initial experiment. This model served as a starting point to understand the dataset and identify the shortcomings of such a minimalistic approach.

#### Results:

Training Accuracy: 25%
Validation Accuracy: 19%
The model's performance was poor, hovering around 25%, which indicated that it was essentially guessing. I realized that I needed a more complex model architecture, especially using convolutional layers, to handle the spatial structure of image data.

## Version 2: Introducing Convolutional Layers
Next, I introduced a convolutional layer with 32 filters to help the model learn spatial features from the images. This marked the transition from a basic dense network to a convolutional neural network (CNN), better suited for image data.

#### Results:

Training Accuracy: 52%
Validation Accuracy: 44%
With the addition of this single convolutional layer, the model's performance improved significantly. However, it was clear that further complexity was needed for the model to generalize better and extract more intricate patterns from the images.

## Version 3: Augmentation Introduction
At this point, I noticed that the model was overfitting the training data, performing better on training than on validation data. To mitigate this, I introduced data augmentation. This allows the model to generalize better by learning from varied versions of the same images.

Training Augmentation: The images were randomly rotated, shifted, zoomed, and flipped, introducing more diversity to the dataset.
Validation & Test Set: No augmentation was applied, as it is typically used only during training to prevent overfitting.
#### Results:

Training Accuracy: 70%
Validation Accuracy: 68%
The use of data augmentation helped reduce overfitting and improved the validation accuracy, making the model more robust when exposed to new data.

## Version 4: Increasing Model Complexity (Adding Layers and Filters)
To further enhance the model’s performance, I decided to increase the depth of the network by adding more convolutional layers with 64 and 128 filters. This allowed the model to capture both lower-level features (such as edges) and higher-level patterns (such as textures) at different layers of abstraction.

#### Results:

Training Accuracy: 85%
Validation Accuracy: 81%
The additional layers significantly improved the model’s ability to detect complex features in the images. The deeper architecture captured more sophisticated patterns and improved generalization to unseen data.

## Version 5: Final Model (Full Architecture)
Finally, I added another convolutional layer with 256 filters and increased the number of neurons in the fully connected (dense) layers to 128 and 64. I also kept data augmentation in place, as it had proven effective in previous iterations.

#### Final Results:

Training Accuracy: 98%
Validation Accuracy: 95%
In this final iteration, the model achieved high accuracy on both the training and validation sets. The combination of a deeper network and data augmentation allowed the model to generalize well to new data and detect complex features in the kidney disease images.

Explanation of Key Decisions and Parameters
**Introducing Data Augmentation (Step 3)**: I decided to use data augmentation after observing overfitting, where the model performed significantly better on training data than on validation data. Augmentation was used to introduce more variety into the training set and prevent the model from memorizing the images.

**32, 64, 128, 256 Filters**: These numbers represent the number of filters in each convolutional layer. Starting with 32 filters helps detect simple features like edges. As the number of filters increases, the network can detect more abstract patterns and higher-level features.

**Max Pooling (2x2)**: Max pooling reduces the spatial dimensions of the feature maps, helping to lower computational costs and prevent overfitting by focusing on the most prominent features in each feature map.

**ReLU Activation**: ReLU introduces non-linearity, allowing the network to learn complex patterns that linear activation functions cannot capture.

**Dense Layers (128, 64 Neurons)**: These fully connected layers help the network combine learned features to make more accurate predictions, acting as a decision-making mechanism after the convolutional layers have extracted meaningful features.

**Use of Categorical Crossentropy and Softmax**: Since this is a multi-class classification problem (with 4 output classes), I used categorical cross-entropy as the loss function, with the softmax function to output class probabilities.
